#### 🧰[***sklearn.decomposition.PCA***](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html?highlight=pca#sklearn.decomposition.PCA)
```python
class sklearn.decomposition.PCA(n_components=None, *, copy=True, whiten=False, svd_solver='auto', tol=0.0, 
iterated_power='auto', n_oversamples=10, power_iteration_normalizer='auto', random_state=None)
```
- 作用：主成分分析   
  (Principal component analysis (PCA).)
- Parameters：
  - ***n_components:int, float or ‘mle’, default=None***
    - 要保留的组件数。  
      如果未设置n_components，则保留所有组件：  
      `n_components == min(n_samples, n_features)`
      - 如果`n_components == 'mle'`和`svd_solver == 'full'`，Minka的MLE用于猜测维度。  
        使用`n_components == 'mle'`会将`svd_solver == 'auto'`解释为`svd_solver == 'full'`。
      - 如果`0 < n_components < 1`和`svd_solver == 'full'`，选择成分的数量，使得需要解释的方差量大于n分量指定的百分比。
      - 如果`svd_solver == 'arpack'`，成分的数量必须严格小于nfeatures和nsamples的最小值。

      因此，None情况会导致：  
      `n_components == min(n_samples, n_features) - 1`
  - ***copy:bool, default=True***
    - 如果为`False`，传递给$fit$的数据将覆盖并运行$fit(X)$。$transform(X)$不会产生预期结果，请改用$fit_transform(X)$。
  - ***whiten:bool, default=False***
    - 当为`True`时，*components_* 矢量乘以 *n_samples* 的平方根，然后除以奇异值，以确保单位分量方差的不相关输出。  
      白化将从变换后的信号中去除一些信息（分量的相对方差尺度），但有时可以通过使下游估计器的数据遵守一些硬连线假设来提高其预测精度。
  - ***svd_solver:{‘auto’, ‘full’, ‘arpack’, ‘randomized’}, default=’auto’***  
    - `auto`:  
      解算器由基于 *X.shape* 和 *n_components* 的默认策略选择：  
      如果输入数据大于500x500，且要提取的组件数量低于数据最小维度的80%，则启用更有效的“随机化”方法。  
      否则，将计算精确的完整SVD，并可选地在之后截断。
    - `full`:  
      通过scipy.linalg.SVD调用标准LAPACK解算器，运行完全的SVD，并通过后处理选择成分。
    - `arpack`:  
      运行SVD截断为 *n_components*，通过scipy.sparse.linalg.svds调用ARPACK解算器。  
      它严格要求 $0<n\_components<min(X.shape)$
    - `randomized`:  
      通过Halko等人的方法运行随机SVD。
  
  - ***tol:float, default=0.0***
    - `svd_solver=='arpack'`计算的奇异值容忍。必须在 $[0.0,\infty)$ 范围内。
  - ***iterated_power:int or ‘auto’, default=’auto’***
    - `svd_solver == 'randomized'`计算的幂法迭代次数。必须在 $[0,\infty)$ 范围内。
  - ***n_oversamples:int, default=10***
    - 仅当`svd_solver == 'randomized'`时，此参数才相关。  
      它对应于额外数量的随机向量来采样 `X` 的范围，以确保适当的调节。
  - ***power_iteration_normalizer:{‘auto’, ‘QR’, ‘LU’, ‘none’}, default=’auto’***
    - 随机SVD求解器的幂迭代规范化器。ARPACK未使用。
  - ***random_state:int, RandomState instance or None, default=None***
    - 当使用 `arpack` 或 `randomized` 解算器时使用。在多个函数调用中传递一个int以获得可重复的结果。
- Attributes：
  - ***components_:ndarray of shape (n_components, n_features)***
  - ***explained_variance_:ndarray of shape (n_components,)***
  - ***explained_variance_ratio_:ndarray of shape (n_components,)***
  - ***singular_values_:ndarray of shape (n_components,)***
  - ***mean_:ndarray of shape (n_features,)***
  - ***n_components_:int***
  - ***n_features_:int***
  - ***n_samples_:int***
  - ***noise_variance_:float***
  - ***n_features_in_:int***
  - ***feature_names_in_:ndarray of shape (n_features_in_,)***

````python
import numpy as np
from sklearn.decomposition import PCA
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
pca = PCA(n_components=1)
pca.fit_transform(X)
````

````python
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
````

#### ***The Iris Dataset***
这些数据集由3种不同类型的虹膜（Setosa、Versicolour和Virginica）花瓣和萼片长度组成，存储在一个150x4的numpy.narray中

- 🗃️行是：**样本**。
- 🌺列是：**萼片长度、萼片宽度、花瓣长度和花瓣宽度**。

下图使用了前两个特征。

````python
# Code source: Gaël Varoquaux
# Modified for documentation by Jaques Grobler
# License: BSD 3 clause

import matplotlib.pyplot as plt

# unused but required import for doing 3d projections with matplotlib < 3.2
import mpl_toolkits.mplot3d  # noqa: F401

from sklearn import datasets
from sklearn.decomposition import PCA

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

plt.figure(1, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
# plt.xticks(())
# plt.yticks(())

````

````python
# To getter a better understanding of interaction of the dimensions
# plot the first three PCA dimensions
fig = plt.figure(2, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)

ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
````

#### ***PCA example with Iris Data-set***  
应用于虹膜数据集的主成分分析。

````python
# Code source: Gaël Varoquaux
# License: BSD 3 clause

import numpy as np
import matplotlib.pyplot as plt


from sklearn import decomposition
from sklearn import datasets

np.random.seed(5)

iris = datasets.load_iris()
X = iris.data
y = iris.target

fig = plt.figure(1, figsize=(4, 3))
plt.clf()

ax = fig.add_subplot(111, projection="3d", elev=48, azim=134)
ax.set_position([0, 0, 0.95, 1])


plt.cla()
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)

for name, label in [("Setosa", 0), ("Versicolour", 1), ("Virginica", 2)]:
    ax.text3D(
        X[y == label, 0].mean(),
        X[y == label, 1].mean() + 1.5,
        X[y == label, 2].mean(),
        name,
        horizontalalignment="center",
        bbox=dict(alpha=0.5, edgecolor="w", facecolor="w"),
    )
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.nipy_spectral, edgecolor="k")

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
````
